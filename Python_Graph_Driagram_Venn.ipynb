```python
from matplotlib_venn import venn3,venn3_unweighted
items=[80,43,6,62,16,3,18]   #[A,B,ab,C,ac,bc,abc]
labels=['Set1','Set2','Set3']
venn3(subsets=items, set_labels=labels,alpha=0.7)  #,ser_colors('red','green','blue'),alpha=0.7
plt.title('3 Circle Venn Diagram')
plt.show()
```
